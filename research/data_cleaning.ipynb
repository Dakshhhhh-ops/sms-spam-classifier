```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("spam.csv", encoding="latin-1")
```

```python
df.drop(
    columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],
    inplace=True
)
```

```python
df.drop_duplicates(inplace=True)
```
